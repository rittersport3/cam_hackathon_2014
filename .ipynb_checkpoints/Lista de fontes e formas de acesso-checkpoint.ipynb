Lista de fontes de dados para o projeto do Hackathon e pequenos scripts de como acessá-los
===

Fontes encontradas:

1. Proposicoes
    1. **ListarProposições** -- comando para entregar todas as proposições por meio de um get/post
    2. **listarProposicoesTramitadasNoPeriodo** -- listar o código das proposições entre um espaço de tempo... talvez interessante para o nosso caso, dado que só queremos apreender o que está acontecendo diáriamente
    3. **ObterProposicaoPorID** -- acesso de fato à proposição, ainda que o texto, de fato, só vai aparecer no tag "LinkInteiroTeor" em um PDF (PQP)
  
2. Discursos no plenário
    1. **ListarDiscursosPlenario** -- Lista discursos... 
    2. **ObterInteiroTeorDiscursosPlenario**  -- Acessa conteúdo em RTF na TAG discursoRTFBase64 
    
Importante: Todos os parametros devem ser enviados no GET e POST, mesmo que com string vazio, necessitam estar lá
==

```python
"""

Modulo de utilidades que obtem os dados 

"""

import requests

ListarProposicoes = {
    "sigla": "",
    "numero": "",
    "ano": "",
    "datApresentacaoIni": "",
    "datApresentacaoFim": "",
    "idTipoAutor": "",
    "parteNomeAutor": "",
    "siglaPartidoAutor": "",
    "siglaUFAutor": "",
    "generoAutor": "",
    "codEstado": "",
    "codOrgaoEstado": "",
    "emTramitacao": ""
}

ObterProposicaoPorID = {
    "idProp": ""
}

ListarProposicoesTramitadasNoPeriodo = {
    "dtInicio": "",
    "dtFim": "",
}


ListarDiscursosPlenario = {
    "dataIni": "",
    "dataFim": "",
    "codigoSessao": "",
    "parteNomeParlamentar": "",
    "siglaPartido": "",
    "siglaUF": "",
}


obterInteiroTeorDiscursosPlenario = {
    "codSessao": "",
    "numOrador": "",
    "numQuarto": "",
    "numInsercao": "",
}

URL = {
    'URL_PROP': {
        'url': 'http://www.camara.gov.br/SitCamaraWS/Proposicoes.asmx/',
        'query': {
            'l_prop': ['ListarProposicoes?', ListarProposicoes],
            'l_tramitadas': ['ListarProposicoesTramitadasNoPeriodo?',
                             ListarProposicoesTramitadasNoPeriodo],

            'get_prop_id': ['ObterProposicaoPorID?', ObterProposicaoPorID]}},
    'URL_SESS': {
        'url': 'http://www.camara.gov.br/SitCamaraWS/SessoesReunioes.asmx/',
        'query': {
            'list_disc': ['ListarDiscursosPlenario?', ListarDiscursosPlenario],
            'get_disc': ['ObterInteiroTeorDiscursosPlenario?',
                         obterInteiroTeorDiscursosPlenario]
            }
        }
    }

            


def get_camara(url,query_op,default_dict, **kwargs):
    default_dict.update(kwargs)
    r = requests.get(url+query_op, params=default_dict)
    return r.content 
```

```python
from lxml import etree  

resp = get_camara(URL["URL_SESS"]["url"],
           URL["URL_SESS"]["query"]["list_disc"][0],
           URL["URL_SESS"]["query"]["list_disc"][1],
           dataIni="01/01/2014",
           dataFim="01/11/2014")

root = etree.XML(resp)

root
```
